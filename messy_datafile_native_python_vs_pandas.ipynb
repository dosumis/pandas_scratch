```python
import re
import pandas as pd
```

```python
# simple python soluton
%time
u = open('./university_towns.txt')
out = []
for l  in u.readlines():
    m = re.match('^(.+)\[edit\]$', l)
    if m:
        state = m.group(1)
        continue
    out.append({'state': state, 'region_name': l.split()[0]})
df = pd.DataFrame.from_records(out)
df[0:10]
```

```python
# Pandas only Solution
# from https://stackoverflow.com/questions/41386443/create-pandas-dataframe-from-txt-file-with-specific-pattern

%time
df = pd.read_csv('./university_towns.txt', sep=";", names=['Region Name']) # Irrelevant sep
df.insert(0, 'State', df['Region Name'].str.extract('(.*)\[edit\]', expand=False).ffill()) # extract
df['Region Name'] = df['Region Name'].str.replace(r' \(.+$', '') # Clean up region name
df = df[~df['Region Name'].str.contains('\[edit\]')].reset_index(drop=True) # clean up state
df[0:10]
```
